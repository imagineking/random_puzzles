```python
import itertools
import numpy 

## only 5 eligible numbers have factors of 11, and 7. They must be separeted. 
## 8 numbers with a factor of 5, and 50 with a factor of 25. total combo (8 choose 2) + 1
# generate a list of numbers with 11 and 7 + (set of (8 choose 2) or 50) + fill in remaining eligible number
# final_list_temp is a list of all the combinations of the 5 numbers that satisify the condition

good_number_list = [6,10,12,14,15,18,20,21,22,24,28,30,33,36,40,42,44,45,48,50,54,55,56,60,66]
eleven_list = [22,33,44,55,66]
seven_list = [14,21,28,42,56]

five_set = list(itertools.combinations([10,15,20,30,40,45,55,60],2))

final_list =[]
for i in range(len(eleven_list)):
    for y in range(len(seven_list)):
        for z in range(len(five_set)):
            final_list.append([eleven_list[i],seven_list[y],five_set[z][0],five_set[z][1]])

for i in range(len(final_list)):
    last_value = 19958400/numpy.prod(final_list[i])
    if last_value not in good_number_list:
        pass
    else:
        final_list[i].append(last_value)

final_list_temp=[]
for i in range(len(final_list)):
    if len(final_list[i])==5:
        final_list_temp.append(final_list[i])
        
aval_number = list(set (good_number_list)-set(eleven_list)-set(seven_list)-{50})
final_list_50=[]
for i in range(len(eleven_list)):
    for y in range(len(seven_list)):
        for x in range(len(aval_number)):
            fifth_number = 19958400/(eleven_list[i]*seven_list[y]*50*aval_number[x])
            if fifth_number == aval_number[x]:
                pass
            elif fifth_number not in aval_number:
                    pass
            else:
                final_list_50.append([eleven_list[i],seven_list[y],50,aval_number[x],fifth_number])

for i in range(len(final_list_50)):
    final_list_50[i].sort()
final_list_50.sort()

final_list_50 = list(final_list_50 for final_list_50,_ in itertools.groupby(final_list_50))
final_list_temp = final_list_temp + final_list_50
```

```python
## Find 2 lists within final_list_temp without dups

global_final = []
for i in range(len(final_list_temp)):
    for y in range(i,len(final_list_temp)):
        if len(set(final_list_temp[i]).intersection(final_list_temp[y])) == 0:
            global_final.append([final_list_temp[i],final_list_temp[y]])
            
## combine 2 lists of global_final without dups, now we have list of 4 possible numbers.
global_final_4 = []
for i in range(len(global_final)):
    for y in range(i,len(global_final)):
        if len(set(global_final[i][0]).intersection(global_final[y][0])) != 0:
            pass
        elif len(set(global_final[i][0]).intersection(global_final[y][1])) != 0:
            pass
        elif len(set(global_final[i][1]).intersection(global_final[y][0])) != 0:
            pass
        elif len(set(global_final[i][1]).intersection(global_final[y][1])) != 0:
            pass
        else:
            global_final_4.append(global_final[i]+global_final[y])
```

```python
## add the remaining numbers, sort & remove dups

for i in range(len(global_final_4)):
    last_five = list(set (good_number_list) - set(list(itertools.chain(*global_final_4[i]))))
    if len(last_five) != 5 or numpy.prod(last_five) != 19958400:
        pass
    else:
        global_final_4[i].append(last_five)

global_final_4_sorted=[]
for i in range(len(global_final_4)):
    global_final_4[i][0].sort()
    global_final_4[i][1].sort()
    global_final_4[i][2].sort()
    global_final_4[i][3].sort()
    global_final_4[i][4].sort()
    global_final_4[i].sort()

global_final_4.sort()
global_final_4_sorted = list(global_final_4 for global_final_4,_ in itertools.groupby(global_final_4))
```
